```python
df['Time'] = pd.to_datetime(df['Time'])
df.sort_values(by="Time", inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
euro_to_drachma = df[['Time','Greek drachma']].copy()
euro_to_drachma['Greek drachma'].value_counts()
```

```python
euro_to_drachma = euro_to_drachma[euro_to_drachma['Greek drachma'] != '-']
euro_to_drachma['Greek drachma'] = euro_to_drachma['Greek drachma'].astype(float)
euro_to_drachma.info()
```

### Visualizations

```python
sns.set_theme()
ax = sns.lineplot(x = euro_to_drachma['Time'],y = euro_to_drachma['Greek drachma'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, size=10)
```

```python
ax = sns.lineplot(x = euro_to_drachma['Time'],y = euro_to_drachma['Greek drachma'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, size=10) 
ax.set(title='Original values')
fig, axes = plt.subplots(3, 2)

sns.lineplot(x = euro_to_drachma['Time'],y = euro_to_drachma['Greek drachma'], ax=axes[0,0]).set(title='Original values')
sns.lineplot(x = euro_to_drachma['Time'],y = euro_to_drachma['Greek drachma'].rolling(7).mean(), ax=axes[0,1]).set(title='Rolling Window:7')
sns.lineplot(x = euro_to_drachma['Time'],y = euro_to_drachma['Greek drachma'].rolling(30).mean(), ax=axes[1,0]).set(title='Rolling Window:30')
sns.lineplot(x = euro_to_drachma['Time'],y = euro_to_drachma['Greek drachma'].rolling(50).mean(), ax=axes[1,1]).set(title='Rolling Window:50')
sns.lineplot(x = euro_to_drachma['Time'],y = euro_to_drachma['Greek drachma'].rolling(100).mean(), ax=axes[2,0]).set(title='Rolling Window:100')
sns.lineplot(x = euro_to_drachma['Time'],y = euro_to_drachma['Greek drachma'].rolling(365).mean(), ax=axes[2,1]).set(title='Rolling Window:365')
```

```python
euro_to_drachma['rolling_mean'] = euro_to_drachma['Greek drachma'].rolling(30).mean()
euro_to_drachma
```
